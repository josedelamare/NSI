## Introduction : Principe du tri par sélection

* Première étape : A partir d’un tableau non trié de n éléments, on parcourt le tableau, et on permute le plus petit élément avec le premier.
* Deuxième étape : On parcourt le tableau à partir du deuxième élément (d’indice 1), et on permute le plus petit élément restant avec le deuxième.
* .........
* ième étape : On parcourt le tableau à partir du ième élément, et on permute le plus petit élément restant avec le ième.
* .........
* (n − 1)ème et dernière étape : On compare les 2 derniers éléments du tableau, et on les permute si besoin.
